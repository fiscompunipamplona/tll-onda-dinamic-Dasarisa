# Comportamiento Ondulatorio en un Fluido

Con el siguiente codigo es posible visualizar graficamente como se comporta un fluido cuando es perturbado por alguna particula.

La función seno que genera la onda es de la forma:

\begin{equation}
    \zeta_1 = \zeta_0\sin(kr_1)\,
\end{equation}

Donde $k$ representa el número de onda y la longitud de onda $\lambda = 5.0\,$cm y amplitud inicial $\zeta_0 = 1.0\,$cm

La grafica representa un espacio de $1.0m \times  1.0m\$

```python
import math
from pylab import sin, sqrt
from pylab import imshow, show, xlabel, ylabel, linspace, inferno
from numpy import loadtxt
salida=open("ondas_est.txt","w")
x=linspace(0,100,1)
y=linspace(0,100,1)
x1=50
y1=50
landa=5.0
e0=1.0
k=2*math.pi/landa
nx=500
dimx=100
x=dimx/nx
ny=500
dimy=100
y=dimy/ny
for i in range(501):
    for j in range (501):
        r1=(math.sqrt(((x1-(i*x))**2)+((y1-(j*y))**2)))
        e1=e0*(sin(k*r1))
        salida.write("%0.2f " %(e0*(sin(k*r1))))
    salida.write("\n")
salida.close()
data = loadtxt("ondas_est.txt", float)
print(data.shape)
imshow(data)
imshow(data, extent=[0,dimx,0,dimx])
imshow(data, origin="lower")
inferno()
show()
```

A continución se muestra una interferencia de onda de tipo lineal de la forma:
\begin{equation}
    \zeta = \zeta_1 + \zeta_2\, .
\end{equation}
Asumiendo que la segunda onda es del tipo
\begin{equation}
    \zeta_2 = \zeta_0\sin(kr_2)\, ,
\end{equation}

```python
salida=open("ondas_est.txt","w")
x=linspace(0,100,1)
y=linspace(0,100,1)
x1=50
y1=50
x2=100
y2=100
landa=5.0
e0=1.0
k=2*math.pi/landa
nx=500
dimx=100
x=dimx/nx
ny=500
dimy=100
y=dimy/ny
for i in range(501):
    for j in range (501):
        r1=(math.sqrt(((x1-(i*x))**2)+((y1-(j*y))**2)))
        r2=(math.sqrt(((x2-(i*x))**2)+((y2-(j*y))**2)))
        e1=e0*(sin(k*r1))
        e2=e0*(sin(k*r2))
        e=e1+e2
        salida.write("%0.2f " %(e1+e2))
    salida.write("\n")
salida.close()
data = loadtxt("ondas_est.txt", float)
print(data.shape)
imshow(data)
imshow(data, extent=[0,dimx,0,dimx])
imshow(data, origin="lower")
inferno()
show()
```
