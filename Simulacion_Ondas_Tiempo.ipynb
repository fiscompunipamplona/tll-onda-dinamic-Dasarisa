# Comportamiento Ondulatorio en un Fluido a lo largo del tiempo

Con el siguiente codigo es posible visualizar graficamente como se comporta un fluido cuando es perturbado por alguna particula a lo largo del tiempo.

La función seno que genera la onda es de la forma:

\begin{equation}
    \zeta_1 = \zeta_0\sin(kr_1+\omega t),
\end{equation}

Donde $k$ representa el número de onda y la longitud de onda $\lambda = 5.0\,$cm , amplitud incial $\zeta_0 = 1.0\,$cm, tiempo de propagacion de $t=100s$ y una velocidad angular de $\omega=100$rad/s

La grafica representa un espacio de $1.0m \times  1.0m\$

```python
import numpy as np
dat=np.zeros((200,200))
np.savetxt("array.txt",dat,fmt="%s")
```

```python
from pylab import plot, imshow, show, xlabel, ylabel, inferno
from numpy import loadtxt, array, zeros, sqrt, pi, sin, linspace
import numpy as np
from IPython import display as disp
import matplotlib.pyplot as plt
salida2=open("Ondas_Est.dat", "w")
gral=800
glob=100
t=100
w=linspace(0,glob,gral)
z=linspace(0,glob,gral)
X,Y=np.meshgrid(w,z)
dat=np.zeros((gral,gral,t),dtype=float)
lda = 5.0
k = 2.*pi / lda
zeta0 = 1.0 
x0 = 50. 
y0 = 50. 
W = 100
for t in range(0,t):
        r=sqrt((X-x0)**2 + (Y-y0)**2)
        e=zeta0*sin(k*r+W*t)
        plt.gca().cla()
        dat[:,:,t]=e
        imshow(e, origin="lower",extent=[0,glob,0,glob])
        inferno()
        disp.clear_output(wait=True)
        disp.display(plt.gcf())
```

A continucion se muestra una interferencia de onda de tipo lineal de la forma:
\begin{equation}
    \zeta = \zeta_1 + \zeta_2\, .
\end{equation}
Asumiendo que la segunda onda es del tipo
\begin{equation}
    \zeta_2 = \zeta_0\sin(kr_2+\omega t)\, ,
\end{equation}

```python
from pylab import plot, imshow, show, xlabel, ylabel, inferno
from numpy import loadtxt, array, zeros, sqrt, pi, sin, linspace
import numpy as np
from IPython import display as disp
import matplotlib.pyplot as plt
salida2=open("Ondas_Est.dat4", "w")
gral=800
glob=100
t=100
w=linspace(0,glob,gral)
z=linspace(0,glob,gral)
X,Y=np.meshgrid(w,z)
dat=np.zeros((gral,gral,t),dtype=float)
lda = 5.0 
k = 2.*pi / lda
zeta0 = 1.0
x0 = 50. 
y0 = 50. 
x1 = 100.
y1 = 100.
W = 100
for t in range(0,t):
        r1=sqrt((X-x0)**2 + (Y-y0)**2)
        e1=zeta0*sin(k*r1+W*t)
        r2=sqrt((X-x1)**2 + (Y-y1)**2)
        e2=zeta0*sin(k*r2+W*t)
        e21=e1+e2
        plt.gca().cla()
        dat[:,:,t]=e21
        imshow(e21, origin="lower",extent=[0,glob,0,glob])
        inferno()
        disp.clear_output(wait=True)
        disp.display(plt.gcf())
```
